```python
   # program for writing data into the file

# opening file for write operation
filev = open("firstfile.txt", "w")

# writing five sentences into the file
print("Enter any five sentences:")

for x in range(5):
    # input of data from user
    sentence = input()

    # writing sentence into the file (with newline)
    filev.write(sentence + "\n")

    print("----------------------")

# closing the file after loop
filev.close()

print("Data successfully written")
```
